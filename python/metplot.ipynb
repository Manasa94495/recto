```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(0.0,2.0,0.01)
y = 1+np.sin(2*np.pi*x)
plt.plot(x,y)
plt.xlabel("X axis value")
plt.ylabel("Y axis value")
plt.title("Representation of Sine")
```

```python
N = 5
men = (20,35,30,35,21)
women = (25,32,21,38,25)
mens = (2,3,4,6,7)
womens = (3,5,2,3,3)
width = 0.35
i = np.arange(N)
p1 = plt.bar(i,men,width,yerr=mens)
p2 = plt.bar(i,women,width,bottom=men,yerr=womens)
plt.legend((p1[0],p2[0]),('Men','Women'))
```

```python
import numpy as np
fig,a=plt.subplots(2,2)
x=np.arange(1,5)
a[0][0].plot(x,x*x)
a[0][1].plot(x,np.sqrt(x))
a[1][0].plot(x,np.exp(x))
a[1][1].plot(x,np.log10(x))
plt.show
```

```python
import seaborn as sns
t = sns.load_dataset("tips")
t.head()
sns.relplot(x='total_bill',y='tip',hue="smoker",style="size",data=t)
sns.relplot(x='time',y='tip',kind='line',data=t)
```

```python
t = sns.load_dataset("tips")
print(t.head())
sns.boxplot(x='day',y='total_bill',data=t)
sns.violinplot(x='tip',y='total_bill',data=t)
sns.catplot(x='tip',y='total_bill',data=t)
sns.catplot(x='tip',y='total_bill',kind='point',data=t)
sns.catplot(x='tip',kind='count',data=t)
sns.catplot(x='tip',y='total_bill',kind='bar',data=t)
```

```python
sns.swarmplot(x='tip',y='total_bill',color='k',size=3,data=t)
```

```python
c=sns.catplot(x='tip',y='total_bill',kind='violin',data=t)
sns.swarmplot(x='tip',y='total_bill',color='k',size=3,data=t,ax=c.ax)
```
